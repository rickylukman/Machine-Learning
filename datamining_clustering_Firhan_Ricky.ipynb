```
df=pd.read_csv('google_review_ratings.csv')
df=df.rename(columns = {'Category 1':'churches','Category 2':'resorts','Category 3':'beaches','Category 4':'parks','Category 5':'theatres','Category 6':'museums','Category 7':'malls','Category 8':'zoo','Category 9':'restaurants','Category 10':'pubs/bars','Category 11':'local services','Category 12':'burger/pizza shops'
                        ,'Category 13':'hotels/other lodgings','Category 14':'juice bars','Category 15':'art galleries','Category 16':'dance clubs','Category 17':'swimming pools','Category 18':'gyms','Category 19':'bakeries','Category 20':'beauty & spas','Category 21':'cafes','Category 22':'view points','Category 23':'monuments','Category 24':'gardens'})
df.head()
```

```
df.describe()
```

```
# Menghapus kolom yang tidak diperlukan
df = df.drop(["User","Unnamed: 25"], axis = 1)
```

Missing Value

```
import seaborn as sns
ax = sns.heatmap(df.isna())
```

```
# Jumlah data kosong untuk setiap kolom
print(df.isnull().sum())
```

```
# Hapus data yang terdapat missing value
df.dropna(inplace=True)
```

```
import matplotlib as mpl
from matplotlib import pyplot as plt
mpl.rcParams['figure.dpi'] = 100
plt.figure(figsize=(100, 100))
f, axes = plt.subplots(1, 2)
sns.boxplot(y=df["restaurants"], ax=axes[0])
sns.boxplot(y=df["pubs/bars"], ax=axes[1])
plt.subplots_adjust(wspace=1)
```

```
# Korelasi
sns.set_style("whitegrid")
ambil = df.select_dtypes(include = ['float64', 'int64',]).iloc[:, :].corr()
plt.figure(figsize=(20, 20))
ax = sns.heatmap(ambil, vmax=1, square=True,annot=True)
plt.xticks
plt.yticks
```
